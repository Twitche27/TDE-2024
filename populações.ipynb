Leitura do dataset e criação das colunas Total e Média

```python
df = pd.read_csv('tabela6579.csv', skiprows=3, nrows=27, delimiter=';')
pd.set_option('display.float_format', '{:.2f}'.format)
df['Total'] = df.iloc[:, 1:].sum(axis=1)
df['Média'] = df.iloc[:, 1:9].mean(axis=1)
df
```

Obtenção dos sub DataFrame's de cada região

```python
#norte = df.loc[0:6].sort_values(by='Total', ascending=False)
norte = df.loc[0:6].sort_values(by='Média', ascending=False)
norte.reset_index(drop=True, inplace=True)
norte
```

```python
#nordeste = df.loc[7:15].sort_values(by='Total', ascending=False)
nordeste = df.loc[7:15].sort_values(by='Média', ascending=False)
nordeste.reset_index(drop=True, inplace=True)
nordeste
```

```python
#sudeste = df.loc[16:19].sort_values(by='Total', ascending=False)
sudeste = df.loc[16:19].sort_values(by='Média', ascending=False)
sudeste.reset_index(drop=True, inplace=True)
sudeste
```

```python
#sul = df.loc[20:22].sort_values(by='Total', ascending=False)
sul = df.loc[20:22].sort_values(by='Média', ascending=False)
sul.reset_index(drop=True, inplace=True)
sul
```

```python
#centro_oeste = df.loc[23:].sort_values(by='Total', ascending=False)
centro_oeste = df.loc[23:].sort_values(by='Média', ascending=False)
centro_oeste.reset_index(drop=True, inplace=True)
centro_oeste
```

Exportação do estado mais populoso de cada região para o arquivo estados.txt

```python
estados_mais_populosos = [norte.iloc[0,0], nordeste.iloc[0,0], sudeste.iloc[0,0], sul.iloc[0,0], centro_oeste.iloc[0,0]]

f = open("estados.txt", "w+")
f.write(f"{estados_mais_populosos}")
f.close()
```
