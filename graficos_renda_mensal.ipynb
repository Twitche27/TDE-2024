```python
import pandas as pd
```

leitura do dataset e alteração do index para ser a Unidade Federativa

```python
manter = [0, 6, 17, 21, 22, 27]
df = pd.read_csv('Rendimento nominal mensal domiciliar per capita da população residente, por UFs.csv', skiprows = lambda x: x not in manter)
df = df.rename(columns={'Unnamed: 0': 'Unidade Federativa'}).set_index('Unidade Federativa').T
df
```
